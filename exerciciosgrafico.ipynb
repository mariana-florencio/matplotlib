Exercício 1: Colunas Simples (Vendas de Produtos por Mês)

```python
import matplotlib.pyplot as plt

mesesx=['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendasy=[120,150,130,170,160,180]

plt.style.use('ggplot')
plt.bar(mesesx,vendasy)
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.title('Vendas de Produtos por Mês')
plt.show()
```

Exercício 2: Colunas Simples (Notas de Alunos em uma Disciplina)

```python
import matplotlib.pyplot as plt

nomesx=['João', 'Maria', 'Pedro', 'Ana', 'Lucas']
notasy=[8,7.5,9,8.5,7]

plt.bar(nomesx,notasy, color='cyan')
plt.xlabel('Alunos')
plt.ylabel('Notas')
plt.title('Notas de Alunos em uma Disciplina')
plt.show()
```

Exercício 3: Colunas Agrupadas (Comparação de Temperatura Máxima e Mínima por
Estação)

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(4)

ymax=[35,28,20,30]
ymin=[25,18,10,20]

width=0.2
plt.title('Comparação de Temperatura Máxima e Mínima por Estação')
plt.bar(x-0.2,ymax,width, color='red')
plt.bar(x,ymin, width, color='blue')
plt.ylabel('TEMPERATURA (°C)')

plt.xticks(x, ['Verão', 'Outono', 'Inverno', 'Primavera'])
plt.legend(['Temp. Máxima (ºC)', 'Temp. Mínima (ºC)'])
plt.show()
```

Exercício 4: Colunas Agrupadas (Desempenho de Dois Times de Futebol em Partidas de um
Campeonato)


```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(5)

ygols_a=[2,3,1,2,4]
ygols_b=[1,2,0,3,1]

width=0.2
plt.title('Desempenho de Dois Times de Futebol em Partidas de um Campeonato')
plt.bar(x-0.2, ygols_a, width, color='orange')
plt.bar(x, ygols_b, width, color='pink')
plt.ylabel('GOLS')


plt.xticks(x, ['Jogo 1', 'Jogo 2', 'Jogo 3', 'Jogo 4', ' Jogo 5'])
plt.legend(['Gols Time A', 'Gols Time B'])
plt.show()
```

Exercício 5: Coluna Horizontal (Receita Mensal de uma Empresa)

```python
import matplotlib.pyplot as plt

xmeses=['Jan', 'Fev', 'Mar', 'Abr']
yreceita=[200,220,210,230]

plt.style.use('ggplot')

plt.figure(figsize=(14,6))
plt.barh(xmeses, yreceita, height=0.5, color='green')

plt.ylabel('Meses')
plt.xlabel('Receita (milhares de reais)')
plt.title('Receita Mensal de uma Empresa')
plt.show()
```
